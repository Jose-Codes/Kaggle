```python
def create_subplots(data, num_rows:int, columns:list, plot_type:str) -> None:
    fig, axs = plt.subplots(num_rows, 5, figsize=(13, 3*num_rows))
    
    axs = axs.flatten()
    num_columns = len(columns)  # Define num_columns based on the length of the columns list
    
    for i, ax in enumerate(axs):
        if i < num_columns:
            column = columns[i]
            
            if plot_type == 'boxplot':
                sns.boxplot(x='class', y=column, data=data, ax=ax)
            elif plot_type == 'hist':
                sns.histplot(data, x=column, hue='class', element="step", stat="density", common_norm=False, ax=ax)
                
            ax.set_title(column)
            ax.set_ylabel(column)
            ax.set_xlabel('')
        else:
            fig.delaxes(ax)  # remove excess plots
    
    plt.tight_layout()
    plt.show()
```

#### Data Preprocessing

Let's try to standardize using z-score normalization the dataset and see how it looks like.

```python
def normalize_data(data):
    normalized_data = data.copy()

    for column in data.drop(['class'], axis=1).columns:
        normalized_data[column] = (data[column] - data[column].mean()) / data[column].std()
    return normalized_data
```

```python
normalized_data = normalize_data(data=data)
normalized_data.head()
```

```python
create_subplots(data=normalized_data, num_rows=num_rows, columns=normalized_data.columns, plot_type='hist') 
```

```python
create_subplots(data=normalized_data, num_rows=num_rows, columns=normalized_data.columns, plot_type='boxplot') 
```

The distribution is the same but now the values have been scaled which should hopefully lead to more stable training and result in a better model. There is still the problem of outliers, I want to train the models with and without outliers and see how that affects the performance.

```python
def remove_outliers(data):
    # Find outliers using Interquartile Range
    # Calculate the upper and lower limits
    for col in data.columns:
        if col == 'class':
            continue
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5*IQR
        upper = Q3 + 1.5*IQR
         
        # Create arrays of Boolean values indicating the outlier rows
        upper_array = np.where(data[col]>=upper)[0]
        lower_array = np.where(data[col]<=lower)[0]
         
        # Create a mask for the rows that are NOT outliers
        mask = (data[col] >= lower) & (data[col] <= upper)

        # Keep only the rows that are NOT outliers
        data = data[mask]
     
    # Print the new shape of the DataFrame
    print("New Shape: ", data.shape)
    return data
```

```python
data_no_outliers = remove_outliers(data=data)
```

```python
normalized_data_no_outliers = remove_outliers(data=normalized_data)
```

```python
create_subplots(data=normalized_data_no_outliers, num_rows=num_rows, columns=normalized_data.columns, plot_type='boxplot') 
```

```python
create_subplots(data=normalized_data_no_outliers, num_rows=num_rows, columns=normalized_data.columns, plot_type='hist') 
```

Analysis: Although as we can see from the boxplots there are still some outliers, but they are not extreme outliers that are beyond the +/- 1.5*IQR. 

As we can also see from the histogram, the distributions are much cleaner now and we are no longer having distributions that look vertical (except for the class labels) after we removed the outliers.

#### Data Sampling
Recall that the data was not equally balanced between all classes, so let's try some sampling techniques. A note from the lecture information on sampling is that random samping may not work very well with skewed data like what we have in this case, so I will do both random and stratified sampling.

```python
def get_percentage_of_class(data):
    all_classes = data['class'].unique()
    
    for class_name in all_classes:
        print(f"Percentage of {class_name} class in dataset: {data[data['class'] == class_name]['class'].size/data['class'].size*100:0.2f}%")
```

```python
get_percentage_of_class(data=data)
```

```python
def random_sampling(data, sample_size):
    # Sampling without replacement
    return data.sample(n=sample_size, replace=False, random_state=42)
```

```python
def stratified_sampling(data):
    stratified_sample_data = pd.DataFrame()

    grouped_data = data.groupby('class')
    min_sample = min(len(indices) for indices in grouped_data.groups.values())
    
    for name, group in grouped_data:
        stratified_sample_data = pd.concat([stratified_sample_data, group.sample(n=min_sample, random_state=42)])

    return stratified_sample_data
```

```python
random_sample_data = random_sampling(data=normalized_data_no_outliers, sample_size=int(0.2*len(normalized_data)))
get_percentage_of_class(data=random_sample_data)
```

```python
stratified_normalized_sample_data = stratified_sampling(data=normalized_data)
get_percentage_of_class(data=stratified_normalized_sample_data)
```

```python
stratified_normalized_sample_data.shape
```

```python
stratified_normalized_no_outliers_sample_data = stratified_sampling(data=normalized_data_no_outliers)
get_percentage_of_class(data=stratified_normalized_sample_data)
```

```python
stratified_normalized_no_outliers_sample_data.shape
```

Now we have a represented stratified sample of the data with no outliers. However it seems like we do not have very many examples, we are at 2.8% of the original dataset.

Let's graph it and see what it looks like :)

```python
create_subplots(data=stratified_normalized_no_outliers_sample_data, num_rows=num_rows, columns=normalized_data.columns, plot_type='boxplot') 
```

```python
create_subplots(data=stratified_normalized_no_outliers_sample_data, num_rows=num_rows, columns=normalized_data.columns, plot_type='hist') 
```

```python
y = data["class"]
x = data.drop(["class"], axis=1)

X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.20, random_state=42)
logistic_model = LogisticRegression(random_state=0, max_iter=10000).fit(X_train, y_train)
```

```python
X_train_predictions = logistic_model.predict(X_train)
X_train_accuracy = accuracy_score(y_train, X_train_predictions)
print(f"Logistic regression validation Accuracy Score on training set: {X_train_accuracy:0.4f}")

X_val_predictions = logistic_model.predict(X_val)
X_val_accuracy = accuracy_score(y_val, X_val_predictions)
print(f"Logistic regression validation Accuracy Score on validation set: {X_val_accuracy:0.4f}")

# Get probabilities
X_train_probabilities = logistic_model.predict_proba(X_train)
X_val_probabilities = logistic_model.predict_proba(X_val)

log_loss_training_set = log_loss(y_train, X_train_probabilities)
print(f"Logistic regression validation Log Loss on validation set: {log_loss_training_set:0.4f}")

log_loss_validation_set = log_loss(y_val, X_val_probabilities)
print(f"Logistic regression validation Log Loss on validation set: {log_loss_validation_set:0.4f}")

lr_raw_data = {"Model":"Logistic Regression",
           "Model Name":"logistic_regression_raw_data",
           "Notes":"No Preprocessing",
           "Accuracy (Training)": X_train_accuracy,
           "Log Loss (Training)": log_loss_training_set,
           "Accuracy (Validation)": X_val_accuracy,
           "Log Loss (Validation)": log_loss_validation_set}

# As of Pandas 2.0 append is deprecated and we have to use concat now, not a fan
# See this post https://stackoverflow.com/questions/75956209/error-dataframe-object-has-no-attribute-append
models_df = pd.concat([models_df, pd.DataFrame([logistic_regression_raw_data])], ignore_index=True)
models_df
```
